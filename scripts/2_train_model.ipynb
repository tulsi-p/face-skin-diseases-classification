```python
# Import necessary libraries
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten, Dense
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

# # Connect to Google Drive
# from google.colab import drive
# drive.mount('/content/drive')

# Set correct paths to your dataset - Updated paths based on your directory structure
base_dir = '/content/drive/MyDrive/MINI_PROJECT'
data_dir = os.path.join(base_dir, 'data')
train_dir = os.path.join(data_dir, 'train')
val_dir = os.path.join(data_dir, 'val')
test_dir = os.path.join(data_dir, 'test')

# Check if directories exist
print(f"Train directory exists: {os.path.exists(train_dir)}")
print(f"Validation directory exists: {os.path.exists(val_dir)}")
print(f"Test directory exists: {os.path.exists(test_dir)}")

# If directories don't exist, try to find them
if not os.path.exists(train_dir) or not os.path.exists(val_dir) or not os.path.exists(test_dir):
    print("Directories not found. Searching for correct paths...")

    # Try to locate the data directory
    for root, dirs, files in os.walk('/content/drive/MyDrive/MINI_PROJECT'):
        if 'data' in dirs:
            data_dir = os.path.join(root, 'data')
            print(f"Found possible data directory: {data_dir}")

            # Look for train, val, test folders with various capitalizations
            possible_train = [os.path.join(data_dir, d) for d in os.listdir(data_dir)
                              if d.lower() in ['train', 'training']]
            possible_val = [os.path.join(data_dir, d) for d in os.listdir(data_dir)
                            if d.lower() in ['val', 'validation']]
            possible_test = [os.path.join(data_dir, d) for d in os.listdir(data_dir)
                             if d.lower() in ['test', 'testing']]

            if possible_train:
                train_dir = possible_train[0]
                print(f"Found training directory: {train_dir}")
            if possible_val:
                val_dir = possible_val[0]
                print(f"Found validation directory: {val_dir}")
            if possible_test:
                test_dir = possible_test[0]
                print(f"Found test directory: {test_dir}")

            break

# Recheck directories after the search
print(f"Using train directory: {train_dir}, exists: {os.path.exists(train_dir)}")
print(f"Using validation directory: {val_dir}, exists: {os.path.exists(val_dir)}")
print(f"Using test directory: {test_dir}, exists: {os.path.exists(test_dir)}")

# Exit if directories still not found
if not os.path.exists(train_dir) or not os.path.exists(val_dir) or not os.path.exists(test_dir):
    raise FileNotFoundError("Could not find the dataset directories. Please check the correct path.")

# Filter out hidden directories and get actual class names
train_classes = [d for d in os.listdir(train_dir) if not d.startswith('.') and os.path.isdir(os.path.join(train_dir, d))]
val_classes = [d for d in os.listdir(val_dir) if not d.startswith('.') and os.path.isdir(os.path.join(val_dir, d))]
test_classes = [d for d in os.listdir(test_dir) if not d.startswith('.') and os.path.isdir(os.path.join(test_dir, d))]

print(f"Classes in train: {train_classes}, count: {len(train_classes)}")
print(f"Classes in val: {val_classes}, count: {len(val_classes)}")
print(f"Classes in test: {test_classes}, count: {len(test_classes)}")

# Verify all directories have the same classes
if set(train_classes) != set(val_classes) or set(train_classes) != set(test_classes):
    print("WARNING: Class mismatch between directories. This will cause errors.")
    print(f"Classes only in train: {set(train_classes) - set(val_classes) - set(test_classes)}")
    print(f"Classes only in val: {set(val_classes) - set(train_classes) - set(test_classes)}")
    print(f"Classes only in test: {set(test_classes) - set(train_classes) - set(val_classes)}")

# Use the common classes across all directories
skin_classes = sorted(list(set(train_classes).intersection(set(val_classes)).intersection(set(test_classes))))
print(f"Common classes across all directories: {skin_classes}")
print(f"Number of common classes: {len(skin_classes)}")

# Count images in each split
def count_images(directory, class_list):
    total = 0
    class_counts = {}
    for class_name in class_list:
        class_dir = os.path.join(directory, class_name)
        if os.path.isdir(class_dir):
            count = len([f for f in os.listdir(class_dir) if not f.startswith('.')])
            class_counts[class_name] = count
            total += count
    return total, class_counts

train_total, train_counts = count_images(train_dir, skin_classes)
val_total, val_counts = count_images(val_dir, skin_classes)
test_total, test_counts = count_images(test_dir, skin_classes)

print(f"Total training images: {train_total}")
print(f"Total validation images: {val_total}")
print(f"Total test images: {test_total}")

# Visualize class distribution
plt.figure(figsize=(12, 6))
plt.bar(train_counts.keys(), train_counts.values())
plt.xticks(rotation=90)
plt.title('Class Distribution in Training Set')
plt.tight_layout()
plt.show()

# Data augmentation for training
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Just rescaling for validation and test
val_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

# Set image size
img_height, img_width = 224, 224
batch_size = 32

# Create data generators with only the common classes
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=True,
    classes=skin_classes  # Only use common classes
)

validation_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False,
    classes=skin_classes  # Only use common classes
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False,
    classes=skin_classes  # Only use common classes
)

# Get class indices
class_indices = train_generator.class_indices
print("Class indices:", class_indices)
class_names = list(class_indices.keys())
num_classes = len(class_names)
print(f"Number of classes for model: {num_classes}")

# Visualize some sample images
def show_batch(image_batch, label_batch):
    plt.figure(figsize=(10, 10))
    for i in range(min(9, len(image_batch))):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(image_batch[i])
        plt.title(class_names[np.argmax(label_batch[i])])
        plt.axis("off")
    plt.tight_layout()
    plt.show()

# Get a batch of images
images, labels = next(train_generator)
show_batch(images, labels)

# Option 1: Build a CNN from scratch
def build_cnn_model():
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=(img_height, img_width, 3)),
        MaxPooling2D(2, 2),
        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D(2, 2),
        Conv2D(128, (3, 3), activation='relu'),
        MaxPooling2D(2, 2),
        Conv2D(128, (3, 3), activation='relu'),
        MaxPooling2D(2, 2),
        Flatten(),
        Dropout(0.5),
        Dense(512, activation='relu'),
        Dropout(0.3),
        Dense(num_classes, activation='softmax')  # Use the correct number of classes
    ])

    model.compile(
        optimizer=Adam(learning_rate=0.0001),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    return model

# Option 2: Use transfer learning with MobileNetV2
def build_transfer_learning_model():
    # Load the MobileNetV2 model
    base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(img_height, img_width, 3))

    # Freeze the base model
    base_model.trainable = False

    # Add custom layers
    model = Sequential([
        base_model,
        tf.keras.layers.GlobalAveragePooling2D(),
        Dense(256, activation='relu'),
        Dropout(0.3),
        Dense(num_classes, activation='softmax')  # Use the correct number of classes
    ])

    model.compile(
        optimizer=Adam(learning_rate=0.0001),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    return model

# Select which model to use (comment out the one you don't want)
# model = build_cnn_model()
model = build_transfer_learning_model()

# Print model summary
model.summary()

# Set up callbacks for training
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)

# Update the checkpoint path to match the correct directory
checkpoint_filepath = os.path.join(base_dir, 'best_model.h5')
model_checkpoint = ModelCheckpoint(
    filepath=checkpoint_filepath,
    save_best_only=True,
    monitor='val_accuracy',
    mode='max'
)

# Calculate steps correctly
steps_per_epoch = train_generator.samples // batch_size
validation_steps = validation_generator.samples // batch_size

# Ensure at least 1 step
steps_per_epoch = max(1, steps_per_epoch)
validation_steps = max(1, validation_steps)

print(f"Steps per epoch: {steps_per_epoch}")
print(f"Validation steps: {validation_steps}")

# Train the model
epochs = 20
history = model.fit(
    train_generator,
    steps_per_epoch=steps_per_epoch,
    epochs=epochs,
    validation_data=validation_generator,
    validation_steps=validation_steps,
    callbacks=[early_stopping, model_checkpoint]
)

# Plot training history
def plot_training_history(history):
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']

    epochs_range = range(len(acc))

    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, acc, label='Training Accuracy')
    plt.plot(epochs_range, val_acc, label='Validation Accuracy')
    plt.legend()
    plt.title('Training and Validation Accuracy')

    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, label='Training Loss')
    plt.plot(epochs_range, val_loss, label='Validation Loss')
    plt.legend()
    plt.title('Training and Validation Loss')
    plt.tight_layout()
    plt.show()

plot_training_history(history)

# Evaluate on test set
test_steps = max(1, test_generator.samples // batch_size)
test_loss, test_acc = model.evaluate(test_generator, steps=test_steps)
print(f"Test accuracy: {test_acc:.4f}")
print(f"Test loss: {test_loss:.4f}")

# Make predictions on test set
test_generator.reset()
y_pred = model.predict(test_generator, steps=test_steps)
y_pred_classes = np.argmax(y_pred, axis=1)

# Get true labels
y_true = test_generator.classes[:len(y_pred_classes)]  # Ensure same length

# Generate classification report
print("Classification Report:")
print(classification_report(y_true, y_pred_classes, target_names=class_names))

# Confusion matrix
plt.figure(figsize=(12, 10))
cm = confusion_matrix(y_true, y_pred_classes)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.title('Confusion Matrix')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.tight_layout()
plt.show()

# Save the model to the correct directory
model_save_path = os.path.join(base_dir, 'skin_disease_model.h5')
model.save(model_save_path)
print(f"Model saved to {model_save_path}")

# Function to predict on a single image
def predict_image(image_path):
    from tensorflow.keras.preprocessing import image

    # Load and preprocess the image
    img = image.load_img(image_path, target_size=(img_height, img_width))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = img_array / 255.0

    # Make prediction
    prediction = model.predict(img_array)
    predicted_class = np.argmax(prediction, axis=1)[0]

    # Print results
    print(f"Predicted class: {class_names[predicted_class]}")
    for i, class_name in enumerate(class_names):
        print(f"{class_name}: {prediction[0][i]:.4f}")

    # Display the image with prediction
    plt.figure(figsize=(6, 6))
    plt.imshow(img)
    plt.title(f"Predicted: {class_names[predicted_class]}")
    plt.axis('off')
    plt.show()

    return class_names[predicted_class]

# Example: Predict on a sample image
# Uncomment the line below and replace with the path to a test image
# sample_image_path = os.path.join(test_dir, class_names[0], os.listdir(os.path.join(test_dir, class_names[0]))[0])
# predict_image(sample_image_path)
```

```python
# Saving files
# Add these imports at the top if not already imported
from sklearn.metrics import roc_curve, auc, precision_recall_curve, average_precision_score
import json
import matplotlib.pyplot as plt
import os
import numpy as np
from itertools import cycle

# Create directory for storing results
final_result_dir = os.path.join(base_dir, 'final_result')
if not os.path.exists(final_result_dir):
    os.makedirs(final_result_dir)
    print(f"Created final results directory: {final_result_dir}")

# Save training history
def save_training_history(history, save_dir):
    # Convert history object to dictionary if it's not already
    if not isinstance(history, dict):
        history_dict = history.history
    else:
        history_dict = history

    # Save history as JSON
    with open(os.path.join(save_dir, 'training_history.json'), 'w') as f:
        json.dump({k: [float(val) for val in v] for k, v in history_dict.items()}, f)

    # Plot and save accuracy
    plt.figure(figsize=(10, 6))
    plt.plot(history_dict['accuracy'], label='Training Accuracy')
    plt.plot(history_dict['val_accuracy'], label='Validation Accuracy')
    plt.title('Model Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(os.path.join(save_dir, 'model_accuracy.png'), dpi=300)
    plt.close()

    # Plot and save loss
    plt.figure(figsize=(10, 6))
    plt.plot(history_dict['loss'], label='Training Loss')
    plt.plot(history_dict['val_loss'], label='Validation Loss')
    plt.title('Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(os.path.join(save_dir, 'model_loss.png'), dpi=300)
    plt.close()

    print(f"Training history plots saved to {save_dir}")

# Generate and save classification report and confusion matrix
def save_classification_metrics(y_true, y_pred_classes, class_names, save_dir):
    # Get classification report as text
    report = classification_report(y_true, y_pred_classes, target_names=class_names)

    # Save classification report
    with open(os.path.join(save_dir, 'classification_report.txt'), 'w') as f:
        f.write(report)

    # Generate and save confusion matrix
    cm = confusion_matrix(y_true, y_pred_classes)
    plt.figure(figsize=(12, 10))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
    plt.title('Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.tight_layout()
    plt.savefig(os.path.join(save_dir, 'confusion_matrix.png'), dpi=300)
    plt.close()

    # Also save confusion matrix as CSV
    pd.DataFrame(cm, index=class_names, columns=class_names).to_csv(
        os.path.join(save_dir, 'confusion_matrix.csv')
    )

    print(f"Classification metrics saved to {save_dir}")

# Generate and save ROC curves for each class
def save_roc_curves(y_true_onehot, y_pred_probs, class_names, save_dir):
    n_classes = len(class_names)

    # Compute ROC curve and ROC area for each class
    fpr = dict()
    tpr = dict()
    roc_auc = dict()

    # First, convert y_true to one-hot encoding if it's not already
    if len(y_true_onehot.shape) == 1:
        y_true_onehot = tf.keras.utils.to_categorical(y_true_onehot, num_classes=n_classes)

    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_true_onehot[:, i], y_pred_probs[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    # Plot all ROC curves
    plt.figure(figsize=(12, 10))
    colors = cycle(['blue', 'red', 'green', 'yellow', 'purple', 'orange', 'cyan', 'magenta', 'black', 'brown'])

    for i, color in zip(range(n_classes), colors):
        plt.plot(fpr[i], tpr[i], color=color, lw=2,
                 label=f'{class_names[i]} (AUC = {roc_auc[i]:.2f})')

    plt.plot([0, 1], [0, 1], 'k--', lw=2)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curves for All Classes')
    plt.legend(loc="lower right")
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(os.path.join(save_dir, 'roc_curves_all.png'), dpi=300)
    plt.close()

    # Generate individual ROC curves
    for i, class_name in enumerate(class_names):
        plt.figure(figsize=(8, 8))
        plt.plot(fpr[i], tpr[i], lw=2, label=f'AUC = {roc_auc[i]:.2f}')
        plt.plot([0, 1], [0, 1], 'k--', lw=2)
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f'ROC Curve for {class_name}')
        plt.legend(loc="lower right")
        plt.grid(True)
        plt.tight_layout()
        plt.savefig(os.path.join(save_dir, f'roc_curve_{class_name}.png'), dpi=300)
        plt.close()

    # Save AUC values
    auc_df = pd.DataFrame({
        'Class': class_names,
        'AUC': [roc_auc[i] for i in range(n_classes)]
    })
    auc_df.to_csv(os.path.join(save_dir, 'auc_values.csv'), index=False)

    print(f"ROC curves saved to {save_dir}")

# Generate and save Precision-Recall curves for each class
def save_precision_recall_curves(y_true_onehot, y_pred_probs, class_names, save_dir):
    n_classes = len(class_names)

    # First, convert y_true to one-hot encoding if it's not already
    if len(y_true_onehot.shape) == 1:
        y_true_onehot = tf.keras.utils.to_categorical(y_true_onehot, num_classes=n_classes)

    # Compute Precision-Recall and average precision for each class
    precision = dict()
    recall = dict()
    avg_precision = dict()

    for i in range(n_classes):
        precision[i], recall[i], _ = precision_recall_curve(y_true_onehot[:, i], y_pred_probs[:, i])
        avg_precision[i] = average_precision_score(y_true_onehot[:, i], y_pred_probs[:, i])

    # Plot all Precision-Recall curves
    plt.figure(figsize=(12, 10))
    colors = cycle(['blue', 'red', 'green', 'yellow', 'purple', 'orange', 'cyan', 'magenta', 'black', 'brown'])

    for i, color in zip(range(n_classes), colors):
        plt.plot(recall[i], precision[i], color=color, lw=2,
                 label=f'{class_names[i]} (AP = {avg_precision[i]:.2f})')

    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title('Precision-Recall Curves for All Classes')
    plt.legend(loc="lower left")
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(os.path.join(save_dir, 'precision_recall_curves_all.png'), dpi=300)
    plt.close()

    # Generate individual Precision-Recall curves
    for i, class_name in enumerate(class_names):
        plt.figure(figsize=(8, 8))
        plt.plot(recall[i], precision[i], lw=2, label=f'AP = {avg_precision[i]:.2f}')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('Recall')
        plt.ylabel('Precision')
        plt.title(f'Precision-Recall Curve for {class_name}')
        plt.legend(loc="lower left")
        plt.grid(True)
        plt.tight_layout()
        plt.savefig(os.path.join(save_dir, f'precision_recall_curve_{class_name}.png'), dpi=300)
        plt.close()

    # Save Average Precision values
    ap_df = pd.DataFrame({
        'Class': class_names,
        'Average_Precision': [avg_precision[i] for i in range(n_classes)]
    })
    ap_df.to_csv(os.path.join(save_dir, 'average_precision_values.csv'), index=False)

    print(f"Precision-Recall curves saved to {save_dir}")

# Function to generate and save all evaluation metrics
def save_all_evaluation_metrics(model, test_generator, history, save_dir):
    print("Generating and saving evaluation metrics...")

    # Ensure the directory exists
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)

    # Reset the generator to the beginning
    test_generator.reset()

    # Predict on test set
    test_steps = max(1, test_generator.samples // test_generator.batch_size)
    y_pred_probs = model.predict(test_generator, steps=test_steps)
    y_pred_classes = np.argmax(y_pred_probs, axis=1)

    # Get true labels
    y_true = test_generator.classes[:len(y_pred_probs)]

    # Convert true labels to one-hot encoding for ROC and PR curves
    y_true_onehot = tf.keras.utils.to_categorical(y_true, num_classes=len(class_names))

    # Save training metrics and plots
    save_training_history(history, save_dir)

    # Save classification report and confusion matrix
    save_classification_metrics(y_true, y_pred_classes, class_names, save_dir)

    # Save ROC curves
    save_roc_curves(y_true_onehot, y_pred_probs, class_names, save_dir)

    # Save Precision-Recall curves
    save_precision_recall_curves(y_true_onehot, y_pred_probs, class_names, save_dir)

    # Save model performance metrics
    test_loss, test_acc = model.evaluate(test_generator, steps=test_steps)
    metrics_summary = {
        "test_accuracy": float(test_acc),
        "test_loss": float(test_loss),
        "number_of_classes": len(class_names),
        "class_names": class_names,
        "total_test_samples": test_generator.samples,
        "batch_size": test_generator.batch_size
    }

    with open(os.path.join(save_dir, 'model_metrics_summary.json'), 'w') as f:
        json.dump(metrics_summary, f, indent=4)

    print(f"All evaluation metrics saved to {save_dir}")

# Call this function after model training and evaluation
save_all_evaluation_metrics(model, test_generator, history, final_result_dir)

# Additional visualization: Class distribution comparison
def save_class_distribution_comparison(train_dir, val_dir, test_dir, class_names, save_dir):
    train_total, train_counts = count_images(train_dir, class_names)
    val_total, val_counts = count_images(val_dir, class_names)
    test_total, test_counts = count_images(test_dir, class_names)

    # Create DataFrame for easier plotting
    dist_df = pd.DataFrame({
        'Class': class_names,
        'Training': [train_counts.get(cls, 0) for cls in class_names],
        'Validation': [val_counts.get(cls, 0) for cls in class_names],
        'Testing': [test_counts.get(cls, 0) for cls in class_names]
    })

    # Save as CSV
    dist_df.to_csv(os.path.join(save_dir, 'class_distribution.csv'), index=False)

    # Plot
    plt.figure(figsize=(14, 8))
    x = np.arange(len(class_names))
    width = 0.25

    plt.bar(x - width, dist_df['Training'], width, label='Training')
    plt.bar(x, dist_df['Validation'], width, label='Validation')
    plt.bar(x + width, dist_df['Testing'], width, label='Testing')

    plt.xlabel('Classes')
    plt.ylabel('Number of Images')
    plt.title('Class Distribution Across Dataset Splits')
    plt.xticks(x, class_names, rotation=90)
    plt.legend()
    plt.tight_layout()
    plt.grid(True, axis='y')
    plt.savefig(os.path.join(save_dir, 'class_distribution_comparison.png'), dpi=300)
    plt.close()

    print(f"Class distribution comparison saved to {save_dir}")

# Save class distribution comparison
save_class_distribution_comparison(train_dir, val_dir, test_dir, skin_classes, final_result_dir)

# Save model architecture diagram (text representation)
def save_model_architecture(model, save_dir):
    # Save model summary to text file
    with open(os.path.join(save_dir, 'model_architecture.txt'), 'w') as f:
        # Redirect summary output to the file
        model.summary(print_fn=lambda x: f.write(x + '\n'))

    # Save model as image (this requires pydot and graphviz to be installed)
    try:
        from tensorflow.keras.utils import plot_model
        plot_model(model, to_file=os.path.join(save_dir, 'model_architecture.png'),
                   show_shapes=True, show_layer_names=True, show_dtype=True)
        print(f"Model architecture diagram saved to {save_dir}")
    except Exception as e:
        print(f"Could not generate model architecture diagram: {e}")
        print("To generate model diagrams, install: pip install pydot graphviz")

# Save model architecture
save_model_architecture(model, final_result_dir)

# Generate example predictions with confidence scores
def save_example_predictions(model, test_dir, class_names, save_dir, num_examples=5):
    from tensorflow.keras.preprocessing import image

    # Create directory for example predictions
    examples_dir = os.path.join(save_dir, 'example_predictions')
    if not os.path.exists(examples_dir):
        os.makedirs(examples_dir)

    plt.figure(figsize=(15, num_examples * 4))

    count = 0
    example_results = []

    # Try to get samples from each class
    for i, class_name in enumerate(class_names):
        class_dir = os.path.join(test_dir, class_name)
        if not os.path.exists(class_dir):
            continue

        files = [f for f in os.listdir(class_dir) if not f.startswith('.')]
        if not files:
            continue

        # Take the first example from this class
        img_path = os.path.join(class_dir, files[0])

        # Load and preprocess the image
        img = image.load_img(img_path, target_size=(img_height, img_width))
        img_array = image.img_to_array(img)
        img_array = np.expand_dims(img_array, axis=0) / 255.0

        # Make prediction
        prediction = model.predict(img_array)
        predicted_class = np.argmax(prediction, axis=1)[0]
        confidence = prediction[0][predicted_class] * 100

        # Plot the image with prediction
        plt.subplot(num_examples, 2, count*2 + 1)
        plt.imshow(img)
        plt.title(f"True: {class_name}\nPred: {class_names[predicted_class]} ({confidence:.1f}%)")
        plt.axis('off')

        # Plot confidence bar chart for all classes
        plt.subplot(num_examples, 2, count*2 + 2)
        conf_values = prediction[0] * 100
        y_pos = np.arange(len(class_names))
        plt.barh(y_pos, conf_values, color=['green' if i == predicted_class else 'gray' for i in range(len(class_names))])
        plt.yticks(y_pos, class_names)
        plt.xlabel('Confidence (%)')
        plt.title('Prediction Confidence')
        plt.grid(True, axis='x')

        # Store results
        example_results.append({
            'true_class': class_name,
            'predicted_class': class_names[predicted_class],
            'confidence': float(confidence),
            'all_confidences': {cls: float(conf) for cls, conf in zip(class_names, conf_values)}
        })

        count += 1
        if count >= num_examples:
            break

    plt.tight_layout()
    plt.savefig(os.path.join(examples_dir, 'example_predictions.png'), dpi=300)
    plt.close()

    # Save example results to JSON
    with open(os.path.join(examples_dir, 'example_predictions.json'), 'w') as f:
        json.dump(example_results, f, indent=4)

    print(f"Example predictions saved to {examples_dir}")

# Save example predictions
save_example_predictions(model, test_dir, class_names, final_result_dir)

print(f"All results have been saved to the final results directory: {final_result_dir}")
```
